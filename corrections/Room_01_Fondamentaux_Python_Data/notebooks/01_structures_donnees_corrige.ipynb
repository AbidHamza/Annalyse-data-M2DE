## Exercice 1 : Manipulation de listes - Solution


```python
ventes_exercice = [120.50, 45.30, 89.90, 234.00, 67.25]

# 1. Ajouter 150.00 à la liste
ventes_exercice.append(150.00)
print(f"Liste avec nouvelle vente : {ventes_exercice}")

# 2. Trouver la vente maximale
vente_max = max(ventes_exercice)
print(f"Vente maximale : {vente_max}")

# 3. Créer une liste des ventes > 100
ventes_superieures_100 = [v for v in ventes_exercice if v > 100]
print(f"Ventes > 100 : {ventes_superieures_100}")
```

## Exercice 2 : Manipulation de dictionnaires - Solution


```python
transactions = [
    {'produit': 'Laptop', 'prix': 899.99, 'quantite': 2},
    {'produit': 'Souris', 'prix': 25.50, 'quantite': 5},
    {'produit': 'Clavier', 'prix': 79.99, 'quantite': 3}
]

# Pour chaque transaction, ajouter la clé 'montant_total'
for transaction in transactions:
    transaction['montant_total'] = transaction['prix'] * transaction['quantite']

print("Transactions avec montant_total :")
for trans in transactions:
    print(trans)
```

## Exercice 3 : Utiliser les compréhensions - Solution


```python
transactions = [
    {'produit': 'Laptop', 'prix': 899.99, 'quantite': 2},
    {'produit': 'Souris', 'prix': 25.50, 'quantite': 5},
    {'produit': 'Clavier', 'prix': 79.99, 'quantite': 3},
    {'produit': 'USB', 'prix': 15.00, 'quantite': 10}
]

# Créer une liste des montants totaux (prix * quantite) pour les produits > 50
montants_eleves = [t['prix'] * t['quantite'] for t in transactions if t['prix'] > 50]
print(f"Montants totaux pour produits > 50 : {montants_eleves}")
```
