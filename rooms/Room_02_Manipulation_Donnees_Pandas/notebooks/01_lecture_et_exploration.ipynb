# Lecture et exploration de données avec Pandas

Ce notebook présente comment charger et explorer des données tabulaires avec pandas.


## Problème métier : charger des données pour l'analyse

Dans la pratique, vous recevez souvent des données sous forme de fichiers CSV. La première étape consiste à charger ces données dans un format exploitable. Pandas fournit la structure DataFrame, idéale pour travailler avec des données tabulaires.


```python
import pandas as pd

# Charger un fichier CSV
# Note : ajustez le chemin selon votre structure de dossiers
df = pd.read_csv('../../data/ventes_commerces.csv')

# Afficher les premières lignes
print("Premières lignes :")
print(df.head())
```

## Explorer la structure des données

Avant d'analyser les données, il est essentiel de comprendre leur structure : nombre de lignes, colonnes, types de données, valeurs manquantes.


```python
# Informations générales sur le DataFrame
print(f"Nombre de lignes : {len(df)}")
print(f"Nombre de colonnes : {df.shape[1]}")
print(f"\nDimensions : {df.shape}")
print(f"\nNoms des colonnes : {list(df.columns)}")
print(f"\nTypes de données :\n{df.dtypes}")
```

```python
# Statistiques descriptives pour les colonnes numériques
print("Statistiques descriptives :")
print(df.describe())

# Informations sur les valeurs manquantes
print("\nValeurs manquantes par colonne :")
print(df.isnull().sum())
```
