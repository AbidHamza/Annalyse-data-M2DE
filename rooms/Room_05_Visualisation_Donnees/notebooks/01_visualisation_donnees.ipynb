# Visualisation de données

Ce notebook présente les visualisations essentielles avec matplotlib et seaborn.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv('../../data/ventes_commerces.csv')

# Histogramme
plt.figure(figsize=(10, 6))
plt.hist(df['montant_total'], bins=30, edgecolor='black')
plt.xlabel('Montant total')
plt.ylabel('Fréquence')
plt.title('Distribution des montants')
plt.show()

# A COMPLETER: Créez un barplot du CA par catégorie
```
