# Corrélations entre variables

Ce notebook présente comment mesurer et interpréter les corrélations entre variables numériques.


## Le coefficient de corrélation de Pearson

Le coefficient de Pearson mesure la relation linéaire entre deux variables. Il varie entre -1 et 1 :
- 1 : corrélation positive parfaite
- 0 : aucune corrélation linéaire
- -1 : corrélation négative parfaite


```python
import pandas as pd

df = pd.read_csv('../../data/ventes_commerces.csv')

# Corrélation entre prix_unitaire et quantite
correlation = df['prix_unitaire'].corr(df['quantite'])
print(f"Corrélation prix-quantité : {correlation:.3f}")

# Matrice de corrélation
print("\nMatrice de corrélation :")
colonnes_numeriques = ['prix_unitaire', 'quantite', 'montant_total']
matrice_corr = df[colonnes_numeriques].corr()
print(matrice_corr)
```
